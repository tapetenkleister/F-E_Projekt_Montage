```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import os
#img = cv.imread('/home/sissi/workspace/F&E/Test/Real_Pic.png',0)
#img2 = img.copy()
load_image = Image.open('/home/sissi/workspace/F&E/Test/Real_Pic.png')
new_image = load_image.resize((1000, 1000))


image = np.asarray(new_image)
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY) 
image=  image.astype(np.float32)

#template = cv.imread('/home/sissi/workspace/F&E/Test/P_top_3.png',0)

# All the 6 methods for comparison in a list
methods = ['cv.TM_CCOEFF', 'cv.TM_CCOEFF_NORMED', 'cv.TM_CCORR',
            'cv.TM_CCORR_NORMED', 'cv.TM_SQDIFF', 'cv.TM_SQDIFF_NORMED']

path = "/home/sissi/Documents/F&E/F-E_Projekt_Montage/images/templates"
for meth in methods:

    for image_name in os.listdir(path):
        print(image_name)
        load_template = Image.open(path + "/" + image_name)
        new_template = load_template.resize((400, 400))
        template = np.asarray(new_template)
        template = cv.cvtColor(template, cv.COLOR_BGR2GRAY) 
        print(template.shape[::-1])
        w, h = template.shape[::-1]
        plt.imshow(template)
        plt.show()
        

        img = image.copy()
        method = eval(meth)
        # Apply template Matching
        res = cv.matchTemplate(img,template,method)
        
        min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)

        print(min_val, max_val, min_loc, max_loc)
        # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
        if method in [cv.TM_SQDIFF, cv.TM_SQDIFF_NORMED]:
            top_left = min_loc
        else:
            top_left = max_loc
        bottom_right = (top_left[0] + w, top_left[1] + h)
        cv.rectangle(img,top_left, bottom_right, 255, 2)
        plt.subplot(121),plt.imshow(res,cmap = 'gray')
        plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
        plt.subplot(122),plt.imshow(img,cmap = 'gray')
        plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
        plt.suptitle(meth)
        plt.show()
```
