```python
from PIL import Image

def get_tile_colors(filename):
    # Open the image file
    image = Image.open(filename)

    # Get the dimensions of the image
    width, height = image.size

    # Calculate the number of tiles in each dimension
    num_tiles_x = width // 20
    num_tiles_y = height // 20

    # Create an empty matrix to store the tile colors
    tile_colors = [[None] * num_tiles_y for _ in range(num_tiles_x)]

    # Iterate over each tile
    for tile_x in range(num_tiles_x):
        for tile_y in range(num_tiles_y):
            # Calculate the bounding box of the tile
            left = tile_x * 20
            upper = tile_y * 20
            right = left + 20
            lower = upper + 20

            # Extract the tile as a new image
            tile_image = image.crop((left, upper, right, lower))

            # Calculate the average RGB values of the tile
            pixel_data = tile_image.load()
            red_sum = 0
            green_sum = 0
            blue_sum = 0
            for x in range(20):
                for y in range(20):
                    r, g, b = pixel_data[x, y]
                    red_sum += r
                    green_sum += g
                    blue_sum += b
            num_pixels = 20 * 20
            red_avg = red_sum // num_pixels
            green_avg = green_sum // num_pixels
            blue_avg = blue_sum // num_pixels

            # Map the average RGB values to a color string
            if red_avg > 2 * (green_avg + blue_avg):
                color = "red"
            elif green_avg > 2 * (red_avg + blue_avg):
                color = "green"
            elif blue_avg > 2 * (red_avg + green_avg):
                color = "blue"
            else:
                color = "yellow"

            # Store the color in the matrix
            tile_colors[tile_x][tile_y] = color

    return tile_colors
```
